```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Carregar o arquivo CSV
df = pd.read_csv('/home/Mall_Customers.csv')

# Verifique se o DataFrame foi carregado corretamente
print(df.head())

# Suponha que as variáveis escolhidas sejam 'Age' e 'Spending Score (1-100)'
X = df[['Age', 'Spending Score (1-100)']]

# Defina o número de clusters
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(X)

# Adicione os rótulos dos clusters ao DataFrame
df['Cluster'] = kmeans.labels_

# Visualize os clusters
plt.scatter(df['Age'], df['Spending Score (1-100)'], c=df['Cluster'], cmap='viridis')
plt.xlabel('Age')
plt.ylabel('Spending Score (1-100)')
plt.title('Clusterização com 2 variáveis')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Suponha que as variáveis escolhidas sejam 'Age', 'Annual Income (k$)', e 'Spending Score (1-100)'
X = df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]

kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(X)

df['Cluster'] = kmeans.labels_

# Visualize os clusters em 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['Age'], df['Annual Income (k$)'], df['Spending Score (1-100)'], c=df['Cluster'], cmap='viridis')

ax.set_xlabel('Age')
ax.set_ylabel('Annual Income (k$)')
ax.set_zlabel('Spending Score (1-100)')
plt.title('Clusterização com 3 variáveis')
plt.show()
```
